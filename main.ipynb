```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("starbucks.csv")
df.head()
```

```python
#grabbing US only locations
df = df.loc[df["Country"] == "US"]
#grabbing only starbucks (not jambda juice)
df = df.loc[df["Brand"] == "Starbucks"]
#might want to only grab only company owned stores
df = df.loc[df["Ownership Type"] == "Company Owned"]
df = df[["Postcode", "State/Province", "City", "Street Address"]]
df = df.dropna()
#grabbing only the first 5 digits of the zipcode
df["Postcode"] = df["Postcode"].apply(lambda s: s[:5])
print(df.shape)
df.head()
```
