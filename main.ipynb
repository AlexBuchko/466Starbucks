Loading income data

```python
incomeDF = pd.read_csv("kaggle_income.csv", encoding="ISO-8859-1")
incomeDF.head()
```

```python
incomeRelevantColumns = [
    "Zip_Code",
    "Median",
    "Mean"
]
incomeDF = incomeDF[incomeRelevantColumns]
incomeDF
```

```python
allJoined = pd.merge(joined, incomeDF, how="left", left_on="zipcode", right_on='Zip_Code')
allJoined.drop(columns="Zip_Code", inplace=True)
allJoined = allJoined.dropna()
allJoined = allJoined.rename(columns={"Median": "median_income", "Mean": "mean_income"})
```

```python
allJoined.to_csv("allJoined.csv")
```
